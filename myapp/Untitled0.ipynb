```python
import ccxt
import pandas as pd
import matplotlib.pyplot as plt

#  API  
api_key = 'your_api_key'
api_secret = 'your_api_secret'

#  BTC    
symbol = 'BTC/USDT'
timeframe = '1d'

#  API  
exchange = ccxt.binance()

#  BTC    
ohlcv = exchange.fetch_ohlcv(symbol=symbol, timeframe=timeframe)

#  pandas DataFrame 
df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df.set_index('timestamp', inplace=True)

#   
period = 20
std_dev = 2
df['middle'] = df['close'].rolling(period).mean()
df['std'] = df['close'].rolling(period).std()
df['upper'] = df['middle'] + std_dev * df['std']
df['lower'] = df['middle'] - std_dev * df['std']

#  
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(df.index, df['close'], label='Close')
ax.plot(df.index, df['upper'], label='Upper')
ax.plot(df.index, df['middle'], label='Middle')
ax.plot(df.index, df['lower'], label='Lower')
ax.legend()
plt.show()
```
