```python
# prompt:    for   6 

import matplotlib.pyplot as plt

x = range(-10, 11)

for i in range(6):
  y = [i**2 + j for j in x]
  plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphs of y = x^2 + i")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#v_wind   .
#a_turbine   .

density = 1.25
PI = 3.14

for i in range(5):
  def f(v_wind):
    a_turbine = 20 + 20*i
    e_mech = (v_wind**2)/2
    W_dot = density*v_wind*(a_turbine**2)/4*PI*0.3*e_mech
    return W_dot
  x = np.linspace(5, 20, 5)
  y = f(x)
  plt.plot(x, y)


plt.xlabel('wind velocity')
plt.ylabel('Energy generated per hour')
plt.show()
```
