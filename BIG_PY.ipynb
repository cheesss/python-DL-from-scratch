```python
print("몇 종류의 주식을 갖고있나요?")
num = int(input())
i = 1
all = 0
while i<=num:
  print(i,"번째 종목은 몇개 갖고있나요?")
  m = int(input())
  
  print(i, "번째 주식은 몇퍼센트 떨어졌나요?")
  p = float(input())

  print("이 주식은 한주에 얼마인가요?")
  v = float(input())
  all = all + v*m*((100-p)/100)
  i = i+1

print("보유한 주식 총액은 ",all,"입니다.")
```
